```python
import math
import numpy as np
import matplotlib.pyplot as plt
from data_splitter import DataSplitter
```

```python
N = 50000
classes = 100
X = np.zeros(N)
Y = [math.floor(np.random.rand()*100) for i in range(N)]
dataset = np.stack([X, Y]).T
```

```python
def plot_size_distr(data):
    plt.hist([np.array(x).shape[0] for x in data])
    plt.xlabel('size')
    plt.ylabel('clients')
    plt.title('Size distribution')
    plt.show()

def plot_label_distr(data):
    plt.hist([np.unique(np.array(x)[:, 1]).shape[0] if np.array(x).shape[0] else 0 for x in client_datasets[:-1]])
    plt.xlabel('size')
    plt.ylabel('clients')
    plt.title('Label distribution')
    plt.show()
```

### IID

```python
data_split_params = {
    'K': 100,
    'split_method': 'iid'
}

data_splitter = DataSplitter(data_split_params, dataset)
client_datasets = data_splitter.split()
```

```python
plot_size_distr(client_datasets)
```

```python
plot_label_distr(client_datasets)
```
