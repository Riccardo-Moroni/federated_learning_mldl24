```python
import matplotlib.pyplot as plt
import numpy as np
from client_selector import ClientSelector
```

```python
def plot_participation_distr(cs):
    res = []
    for i in range(10000):
        res.append(cs.sample())
    res = np.array(res).flatten()
    _, counts = np.unique(res, return_counts=True)
    plt.hist(counts)
    plt.xlabel('Times selected')
    plt.ylabel('Clients num')
    plt.title('Participation')
    plt.show()


def plot_client_distr(cs):
    res = []
    for i in range(10000):
        res.append(cs.sample())
    res = np.array(res).flatten()
    cl, counts = np.unique(res, return_counts=True)
    ids = range(100)
    vals = np.array([counts[np.where(cl == i)[0][0]] if (i in cl) else 0 for i in ids])
    plt.bar(ids, vals)
    plt.xlabel('Client')
    plt.title('Participation per client')
    plt.show()
```

### Uniform

```python
params = {
    'K': 100,
    'C': 0.1,
    'participation': 'uniform'
}

cs = ClientSelector(params)
plot_participation_distr(cs)
```

```python
plot_client_distr(cs)
```

### Skewed

#### $\gamma = 0.1$

```python
params = {
    'K': 100,
    'C': 0.1,
    'participation': 'skewed',
    'gamma': 0.1
}

cs = ClientSelector(params)
plot_participation_distr(cs)
```
